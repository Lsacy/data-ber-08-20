```python
# since the difference are minial, less than 10%, I will drop href
df.drop('href formula',axis=1,inplace=True)
```

```python
# reset the index and drop the old index and the 'original order' column
df = df.drop('original order',axis=1)
df = df.reset_index(drop=True)
```

```python
df.index=range(len(df))
```

```python
df.index.name = 'id'
df
```

```python

# drop [year] column, since [Date] column already contains information on year
df = df.drop('Year', axis=1)
```

```python
# now we examine the fatality by looking close to the fatal column
set(df['Fatal (Y/N)'])
```

```python
# it contains many different values for essentially 3 types of information: yes, no, unknow , 
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].str.strip()
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].fillna('U')
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].str.replace('n', 'N')
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].str.replace('#VALUE!', 'U')
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].str.replace('N', 'N')
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].str.replace('F', 'Y')
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].str.replace('UNKNOWN', 'U')
df.rename(columns={'Fatal (Y/N)': 'Fatal (Y/N/U)'}, inplace=True)
```

```python
set(df['Fatal (Y/N/U)'])
```

```python
# now we check for sex
set(df['Sex'])
```

```python
df.columns
```

```python
df.rename(columns={'Sex ':'Sex'},inplace=True)
df.columns = df.columns.str.strip()
```

```python
set(df['Sex'])
```

```python
df['Sex'] = df['Sex'].str.strip()
df['Sex'] = df['Sex'].fillna('U')
df['Sex'] = df['Sex'].str.replace('N', 'U')
df['Sex'] = df['Sex'].str.replace('lli', 'U')
df['Sex'] = df['Sex'].str.replace('nan', 'U')
df['Sex'] = df['Sex'].str.replace('.', 'U')
```

```python
set(df['Sex'])
```

```python
# look good now, we will fill the rest of the missing values with unknown - U
df.fillna('U')
```

```python
# exporting the table to a new csv file and we are done
df.to_csv('GSF5_clean.csv', index=False)
```
